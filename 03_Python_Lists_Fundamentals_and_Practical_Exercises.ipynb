# Lists in Python


## 01 Exercise: List creation

Problem: Create a list called a_list with the following elements:
1, "hello", [1, 2, 3], True



```python

a_list = [1, "hello", [1, 2, 3], True]

a_list # creating a list with mixed data types
```

In Python, lists can store different data types at the same time.
This list contains an integer, a string, another list (nested list) and a boolean value

This flexibility is one of the main reasons lists are widely used

## 02 Exercise: Accessing an element by index

Problem: Find the value stored at index 1 in a_list

```python

a_list[1]# withg this we access the element at index 1
```

## 03 Exercise: Slicing a list

Problem: Get the elements stored at index 1, 2, and 3 from a_list

```python

a_list[1:4]#slicing the list from index 1 up to (but not including) index 4 since lists follow this format list[start:end]

```

## 04 Exercise: List concatenation

Problem: Concatenate the following lists:
A = [1, "a"] and B = [2, 1, "d"]

```python
A = [1, "a"]
B = [2, 1, "d"]


print(A + B) # concatenating both lists
```

When the + operator is used with lists and it joins them instead of performing a mathematical operation.
The result is a new list that contains all elements from both lists in order
