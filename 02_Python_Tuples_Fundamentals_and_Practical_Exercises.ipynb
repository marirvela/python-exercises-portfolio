# Tuples in Python


## 01 Exercise: Tuple length

Problem: Determine how many elements are stored in the tuple



```python
genres_tuple = (
    "pop",
    "rock",
    "soul",
    "hard rock",
    "soft rock",
    "R&B",
    "progressive rock",
    "disco"
)

len(genres_tuple) # this function returns the number of elements in a tuple
#  this operation does not modify the data it only care about its structure

```

## 02 Exercise: Accessing an element by index

Problem: Access the element located at index 3 in the tuple

```python
genres_tuple[3]
# since the first element starts at index 0 accessing index 3 retrieves the fourth element in the tuple
```

## 03 Exercise: Slicing a tuple

Problem: Obtain the elements at indices 3, 4, and 5 using slicing

```python
genres_tuple[3:6] # in this case we use slicing because it allows us to get a subset of elements without modifying the original tuple
# the slice includes the start index but excludes the end index returning a new tuple
```

## 04 Exercise: Retrieving the first elements

Problem: Extract the first two elements from the tuple

```python
genres_tuple[0:2]
# this retrieves the initial 2 elements of the tuple, it works this way [start, end+1]
```

## 05 Exercise: Finding the index of an element

Problem: Find the index of the value "disco" within the tuple

```python
genres_tuple.index("disco") # this method returns the position of the first occurrence of the specified value
```

## 06 Exercise: Sorting data from a tuple

Problem: Generate a sorted list from the tuple C_tuple = (-5, 1, -3)

```python
C_tuple = (-5, 1, -3)
C_sorted = sorted(C_tuple) # this function returns a new list containing the elements in ascending order
C_sorted
# With this exercise I realized the difference between immutable data storage (tuples) and mutable data manipulation (lists)
```
