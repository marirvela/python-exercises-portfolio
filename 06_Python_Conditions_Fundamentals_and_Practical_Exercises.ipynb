## 01 Exercise: Basic if statement

Problem: Write down an if statement that determines if a musical disc has a rating > 8. Test it using "Back in Black" rating which is 8.5. If the result is "True", print "This album is Amazing!"



```python
back_in_black_rating = 8.5
if back_in_black_rating > 8: #here i use a simple if statement to evaluate a boolean condition
    print ("This album is Amazing!") #Since the rating is 8.5, which is greater than 8, the condition evaluates to true
```

This exercise reinforces how Python makes decisions based on logical comparisons and executes code only when conditions are met

## 02 Exercise: if else decision making

Probem: Write down an if-else statement that if an album's rating is > 8 prints "This album is Amazing!", but if it is <= 8 prints "This album is OK!"

```python
anymusicalalbum_rating = 6
if anymusicalalbum_rating > 8: #if the condition is true, one message is printed
  print("This album is Amazing!")
else: # otherwise python automatically falls back to the alternative path
  print("This album is OK!")
```

This example introduces an else branch, allowing the program to handle both possible outcomes of the condition
Also the structure uused is essential for building logic that reacts to different scenarios instead of assuming only one outcome

## 03 Exercise: compound conditions with logical operators

Problem: Create an if statement that will determine if a musical album was released before 1980 or during 1991 or 1993. If the condition is true it will print the year of release.

```python
year_of_release = 1991

#the album qualifies if any of the conditions is true, not necessarily all of them
if year_of_release < 1980 or year_of_release == 1991 or year_of_release == 1993:
    print ("This album was released on: ", year_of_release)
```

Using compound conditions like this allows more expressive and realistic decision making, similar to real world rules where multiple cases are acceptable
