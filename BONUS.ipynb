#### 1.1. Problem: Write down a function that will return the lesser of two evens, but in case among them there is any odd it will return the greater.

```python
def lesseroftwoevens(a, b): # We define the lesser of two evens function
  if(a%2)==0: # Here we start the first comparisson to know if a is even
    if(b%2)==0: # Here we check if b is also an even
      if a<b: # In case both cases are true, now we will check which is the lesser
        print(a) # If a < b we will print a
        return a

      else:
        print(b) # If not we will print b
        return b

    else:
      if a>b: # Now, in case b is not even it means is odd so the output changes
        print(a) # we will check which is the greater, if a is greater than b we print a
        return a

      else:
        print(b) # if not, it means b is greater than a so we will print b
        return b

  else:
    if a>b: # If none of them are evens, we will print the greater among them
      print(a) # If a is greater than b, we will print a
      return a
    else:
      print(b) # If not we will print b
      return b



```

Trying it out:

```python
lesseroftwoevens(2, 4)
```

```python
lesseroftwoevens(4, 7)
```

#### 1.3. Problem: Given 2 integers, this function will return True if the sum of them is 20. In case is not, it must return False

```python
def daveinte(n1, n2): # Definition of the function
  if n1 == 20: # Evaluates if the first number is 20
    return True # in case it is, we will return true
  else:
    if n2 == 20: # if not, we will evaluate if the second number is 20
      return True # in case it is, we will return true
    else:
      if n1 + n2 == 20: # if not, we will evaluate if the sum of them is 20
        return True # in case it is, we will return true
      else:
        return False # in case it is not, we will return false
```

2.1. Problem: write a function that will capitalize the first and the fourth letter of an string

```python
def uppercase(str): # here we define the function
  first = str[0][0].upper() # with this we capitalize the first letter
  fourth = str[3][0].upper() # and with this we capitalize the fourth letter
  print(first + str[1:3] + fourth + str[4:]) # and with this we concatenate everything up
  return first + str[1:3] + fourth + str[4:]
```

```python
uppercase('holaatodos')
```

2.2. Problem: given a list of integers, return true if it contains a 3 next to a 3 somewhere

```python
def has33(nums):  # here we define the function
    for i in range(len(nums)-1):  # we loop until the second last element to avoid index error
        if nums[i] == 3 and nums[i+1] == 3:  # here we check if current and next value are both 3
            return True  # if we find two 3 together we return True immediately
    return False  # if loop finishes and we never found it, then we return False
```

Trying it out

```python
has33([1, 3, 3])
```

```python
has33([1, 3, 1, 3])
```

2.3. Problem: given a string, return a new string where for every character there are three characters

```python
def triplechar(str):  # here we define the function
    result = ""  # we create an empty string to store the final result
    for char in str:  # we loop through each character in the string
        result += char * 3  # here we repeat each character three times and add it
    return result  # finally we return the modified string
```

Trying it out

```python
triplechar("chocolate")
```

3.1. Problem: given a string, return the number of even integers from 1 to 100 that are divisible by 3

```python
def even_divisible_by_three():  # here we define the function
    count = 0  # we start a counter at 0
    for i in range(1, 101):  # we loop from 1 to 100
        if i % 2 == 0 and i % 3 == 0:  # here we check if number is even and divisible by 3
            count += 1  # if condition is true, we increase the counter
    return count  # return the final count
```

3.2. Problem: wwrite a function that returns a list of words from a sentence that start with "s"

```python
def starts_with_s(sentence):  # define the function
    words = sentence.split()  # we split the sentence into words
    result = []  # empty list to store matching words
    for word in words:  # loop through each word
      if word[0].lower() == "s":  # i check if first letter is s in lower for better results
        result.append(word)  # if yes, add it to the result list
    return result  # okay and we return final list
```
