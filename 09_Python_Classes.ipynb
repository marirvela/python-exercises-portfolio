# Classes in Python


## 1. Working with the Circle class



```python
import matplotlib.pyplot as plt #here we will import some important libraries
%matplotlib inline

class Circle(object): #this will create the Circle class

    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color

    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)

    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()
```

Creating an instance of the Circle class

```python
# Create a Circle object called RedCircle
RedCircle = Circle(10, "red")
# here I create an instance of the Circle class
# I pass the radius and the color because the constructor expects those values
# this is now a real object, not just the class definition
```

Exploring the object using dir()

```python
dir(RedCircle)
# dir() shows all the attributes and methods available for this object
# this includes methods defined in the class and default Python methods
# I use this to understand what I can do with the object
```

Accessing object attributes

```python
RedCircle.radius
# here i access the radius attribute stored inside this specific object

RedCircle.color
# and here this access the color attribute of the object
# each instance keeps its own values
```

Modifying object attributes

```python
RedCircle.radius = 1
# here I directly modify the radius attribute of the object
# this only affects RedCircle, not the Circle class or other objects

RedCircle.radius
# I check the value to confirm that the change was applied
```

Drawing the circle using a method

```python
RedCircle.drawCircle()
# I call the drawCircle() method from the object
# um here the method uses the object's own radius and color internally
```

Increasing the radius using a method

```python
print("Radius of object:", RedCircle.radius)
# here i print the current radius before modifying it

RedCircle.add_radius(2)
# here i increase the radius by 2 using the method
# using a method is cleaner than changing the attribute manually

print("Radius of object after applying the method add_radius(2):", RedCircle.radius)
# we need to check the updated radius

RedCircle.add_radius(5)
# here i increase the radius again by 5

print("Radius of object after applying the method add_radius(5):", RedCircle.radius)
# and finaly confirm the final value
```

Creating another Circle instance (using default values)

```python
BlueCircle = Circle(radius=100)
# Here I create another Circle object
# I only pass the radius and let the color take the default value
# This shows how default parameters work in the constructor
```

Accessing attributes of the new object

```python
BlueCircle.radius
# here i access the radius of the BlueCircle object

BlueCircle.color
# and here i access the color attribute
# even though I did not pass it, the object still has it
```

Drawing the second circle

```python
BlueCircle.drawCircle()
# Draw the BlueCircle object
# Each object draws itself based on its own attributes
```

### **Important observation**
Even though RedCircle and BlueCircle come from the same class,
they do not share attribute values.
Each instance has its own data (like position, radius, and color).

This is important because it would be incorrect to assume that changing one object affects the other.
The instances are independent

Defining the Rectangle class

Defining a method to draw the rectangle

```python
class Rectangle(object):

    def __init__(self, width=2, height=3, color="r"):
        self.height = height
        # Store the height inside the object

        self.width = width
        # Store the width inside the object

        self.color = color
        # Store the color of the rectangle
##Here the constructor defines what every Rectangle object will have.
##Each instance gets its own width, height, and color
    def drawRectangle(self):
        plt.gca().add_patch(
            plt.Rectangle((0, 0), self.width, self.height, fc=self.color)
        )
        # Draw the rectangle using the object's own attributes

        plt.axis('scaled')
        # Keep the proportions correct

        plt.show()
        # Display the rectangle
```

Creating an instance of Rectangle

```python
SkinnyBlueRectangle = Rectangle(2, 10, "blue")
# Create a Rectangle object with specific width, height, and color
# This is a different object from Circle, even though both can be drawn
```

Accessing rectangle attributes

```python
SkinnyBlueRectangle.height
# with this i access the height of the rectangle

SkinnyBlueRectangle.width
# with this i access the width of the rectangle

SkinnyBlueRectangle.color
# and with this i access the color of the rectangle
```

Drawing the rectangle

```python
SkinnyBlueRectangle.drawRectangle()
#here i call the method to draw the rectangle
#and the method uses the values stored inside this object
```
