# Dictionaries in Python


## 01 Exercise: Dictionary keys

Problem: Identify the keys stored in the dictionary soundtrack_dic



```python
soundtrack_dic = {"The Bodyguard":"1992", "Saturday Night Fever":"1977"}# ddictionaries store data as keyvalue pairs
soundtrack_dic
soundtrack_dic.keys()# here keys represent unique identifiers for each value
```

Keys are the main access point of a dictionary. Here, they represent movie titles, which makes sense because each soundtrack is uniquely identified by its name

## 02 Exercise: Dictionary values

Probem: Identify the values stored in the dictionary soundtrack_dic

```python
soundtrack_dic.values()# with values we represent the data associated with each key
# in this case they store the release year of each soundtrack
```

here I would like to embrace that while keys answer “what is it?” values answer “what information is stored about it?” and that is what makes dictionaries so powerful

## 03 Exercise: Creating a dictionary from scratch

Problem: create a dictionary where album names are the keys and sales (in millions) are the values

```python
album_sales_dict = {"The Bodyguard":50, "Back in Black":50, "Thriller":65} #here we create the dictionary to map album names to their sales

album_sales_dict
```

This is a real world use case: fast lookups, clear relationships and meaningful keyvalue pairing

## 04 Exercise: accessing a value by key

Problem: Find the total sales of Thriller using the dictionary

```python
album_sales_dict["Thriller"]
#we could access values in a dictionary is done using the key
```

Here with dictionaries they allow us to have direct access to data that means no indexing, no guessing positions, just clean and explicit retrieval

## 05 Exercise: Retrieving dictionary keys

Problem: obtain the names of all albums stored in the dictionary

```python
album_sales_dict.keys() #here the keys represent the album names stored in the dictionary
#this can be useful when iterating over datasets
```

## 06 Exercise: Retrieving dictionary values

Problem: Retrieve all sales records stored in the dictionary

```python
album_sales_dict.values() #when weuse values wecan access to the actual stored data
```
