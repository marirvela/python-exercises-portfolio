# Functions in Python


## 01 Exercise: Function to divide

Problem: Create a function that divides the first input by the second input



```python
def div(a, b):              # here I define a function because I want to reuse this operation instead of repeating code
    return a / b            # I return the result since I want to use it later, not just print it on the screen
```

This function takes two parameters and returns the result of dividing the first by the second

Using return allows the result to be reused later instead of just printed, which is the correct approach when building reusable logic

## 02 Exercise:function to combine 2 values

Probem: Use a function to add two values

```python
def con(a, b):              # I create this function to combine two values without worrying about the data type yet
    return a + b            # I use + because Python already knows how to handle it depending on the type
```

The logic is wrapped inside a function, making the operation reusable and easier to read wherever it’s needed

## 03 Exercise: using the function with integers

Problem: Can the previously defined function be used to add integers or strings?

```python
con(2, 2)                   # I pass two integers, so Python performs a normal arithmetic addition
```

My response is yes. The same function works with integers without any changes.
The behavior depends on the data type passed to the function not on the function definition itself.

## 04 Exercise: using the same function with lists

Problem: Can the same function be used to concatenate a list or a tuple?

```python
con(['a', 'b'], [1])        # now I pass lists, so + concatenates them instead of adding numbers
```

This works because the + operator behaves differently depending on the data type.
In this case, it concatenates lists instead of doing arithmetic
